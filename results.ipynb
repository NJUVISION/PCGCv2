## BD-Rate

```python
from utils import bdrate_metrics
```

```python
def points_for_bdrate(df, index_rate='bpp', index_psnr='mseF,PSNR (p2point)'):
    rate = df[index_rate]
    psnr = df[index_psnr]
    rd_points = [tuple((a, b)) for a, b in zip(rate, psnr)]
    return rd_points
```

```python
rd_points_gpcc_octree = points_for_bdrate(gpcc_octree, 'bpp', metric)
rd_points_gpcc_trisoup = points_for_bdrate(gpcc_trisoup, 'bpp', metric)
rd_points_vpcc = points_for_bdrate(vpcc, 'geometry bpp', metric)
rd_points_learnedpcgc = points_for_bdrate(learnedpcgc, 'bpp', metric)
rd_points_multiscalepcgc = points_for_bdrate(multiscalepcgc, 'bpp', metric)
```

#### vpcc

```python
bdrate_vpcc = bdrate_metrics.bdrate(rd_points_vpcc, rd_points_multiscalepcgc)
print(round(bdrate_vpcc, 2))
```

#### gpcc octree

```python
bdrate_octree = bdrate_metrics.bdrate(rd_points_gpcc_octree, rd_points_multiscalepcgc)
print(round(bdrate_octree, 2))
```

#### gpcc trisoup

```python
bdrate_trisoup = bdrate_metrics.bdrate(rd_points_gpcc_trisoup, rd_points_multiscalepcgc)
print(round(bdrate_trisoup, 2))
```

#### learnedpcgc

```python
bdrate_learnedpcgc = bdrate_metrics.bdrate(rd_points_learnedpcgc, rd_points_multiscalepcgc)
print(round(bdrate_learnedpcgc, 2))
```
